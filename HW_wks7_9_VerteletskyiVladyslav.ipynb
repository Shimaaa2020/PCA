Here we implement PCA for the wine dataset from sklearn. First, let's import required libraries and functions:

```
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.decomposition import PCA
from sklearn.datasets import load_wine
```

We load predictors as numerical and labels as a categorical variables

```
X, y = load_wine(return_X_y=True)
```

And apply PCA:

```
pca = PCA(n_components=2)
pca.fit(X)
```

Now we project X onto principal components and plot the result colored with respect to a class:

```
X_p = X @ pca.components_.T
plt.scatter(X_p[:,0], X_p[:,1], c=y)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Result of PCA application with n=2')
```
